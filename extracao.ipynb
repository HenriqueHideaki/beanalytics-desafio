#### Historico de preços

```python
import requests
import json

# URL da API do SteamDB
url = 'https://steamdb.info/api/GetPriceHistory/?appid=2390&cc=br'

# Cabeçalho para simular um navegador real
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36',
    'Accept': 'application/json',
    'Accept-Encoding': 'gzip, deflate, br, zstd',
    'Accept-Language': 'pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7',
    'Referer': 'https://steamdb.info/app/2390/',
    'X-Requested-With': 'XMLHttpRequest'
}

# Fazer a requisição GET
response = requests.get(url, headers=headers)

# Verificar o status da resposta
if response.status_code == 200:
    # Verificar o conteúdo da resposta
    content_type = response.headers.get('Content-Type', '')
    print("Content-Type:", content_type)

    # Verificar se o conteúdo é JSON
    if 'application/json' in content_type:
        try:
            data = response.json()
            
            # Verificar se a resposta JSON contém dados
            if data.get("success"):
                history = data["data"]["history"]
                sales = data["data"]["sales"]

                # Imprimir dados do histórico
                print("Histórico de Preços:")
                for item in history:
                    timestamp = item["x"]
                    price = item["y"]
                    formatted_price = item["f"]
                    discount = item["d"]
                    sale_name = sales.get(str(timestamp), "No Sale")

                    print(f"Timestamp: {timestamp}, Price: {price}, Formatted Price: {formatted_price}, Discount: {discount}%, Sale: {sale_name}")
            else:
                print("Falha ao carregar dados do JSON.")
        except json.JSONDecodeError as e:
            print("Erro ao decodificar JSON:", e)
            print("Conteúdo da resposta:", response.text)
    else:
        print("Resposta não está em formato JSON. Conteúdo da resposta:", response.text)
else:
    print(f'Falha ao acessar o SteamDB. Status code: {response.status_code}')
```

```python
import json
from datetime import datetime, timedelta
import pandas as pd
import plotly.express as px

# Carregar o conteúdo JSON do arquivo
with open('graph_followers_response.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extrair os dados da série temporal
start_timestamp = data["data"]["start"]
step_seconds = data["data"]["step"]
values = data["data"]["values"]

# Converter timestamps em datas e criar um DataFrame
dates = [datetime.utcfromtimestamp(start_timestamp + i * step_seconds) for i in range(len(values))]
followers_data = pd.DataFrame({'Date': dates, 'Followers': values})

# Remover valores nulos
followers_data.dropna(inplace=True)

# Definir o índice do DataFrame como a coluna de datas
followers_data.set_index('Date', inplace=True)

# Filtrar os dados para o último ano
one_year_ago = datetime.now() - timedelta(days=365)
filtered_data = followers_data[followers_data.index >= one_year_ago]

# Resample para obter a média de seguidores por mês
monthly_followers = filtered_data.resample('M').mean()

# Reset index para facilitar o uso no Plotly
monthly_followers.reset_index(inplace=True)

# Criar o gráfico com Plotly
fig = px.line(monthly_followers, x='Date', y='Followers', title='Média de Seguidores por Mês (Último Ano)', markers=True)

# Adicionar valores nos pontos
fig.update_traces(text=monthly_followers['Followers'].round(2), textposition="top center")

# Ajustar o layout do gráfico
fig.update_layout(
    xaxis_title='Data',
    yaxis_title='Número de Seguidores',
    yaxis=dict(tickvals=[0, 25000, 50000, 75000, 100000]),
    yaxis_range=[0, 100000],
    hovermode='x unified'
)

# Mostrar o gráfico
fig.show()
```
